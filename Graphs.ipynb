```python
data = sorted(data, key=lambda r: r[2])
P = 256

plt.plot([1000000, 5000000, 10000000, 50000000, 100000000, 500000000], np.log10([
    r[4] for r in data
    if (r[0] == 0 and \
        r[2] in [1000000, 5000000, 10000000, 50000000, 100000000, 500000000] and \
        r[3] == P)
]), label="No B+ Tree")
plt.plot([1000000, 5000000, 10000000, 50000000, 100000000, 500000000], np.log10([
    r[4] for r in data
    if (r[0] == 1 and \
        r[2] in [1000000, 5000000, 10000000, 50000000, 100000000, 500000000] and \
        r[3] == P)
]), label="Simple B+ Tree")
plt.plot([1000000, 5000000, 10000000, 50000000, 100000000, 500000000], np.log10([
    r[4] for r in data
    if (r[0] == 2 and \
        r[1] == 4 and \
        r[2] in [1000000, 5000000, 10000000, 50000000, 100000000, 500000000] and \
        r[3] == P)
]), label="Vector B+ Tree, B=4")
plt.plot([1000000, 5000000, 10000000, 50000000, 100000000, 500000000], np.log10([
    r[4] for r in data
    if (r[0] == 2 and \
        r[1] == 16 and \
        r[2] in [1000000, 5000000, 10000000, 50000000, 100000000, 500000000] and \
        r[3] == P)
]), label="Vector B+ Tree, B=16")
plt.plot([1000000, 5000000, 10000000, 50000000, 100000000, 500000000], np.log10([
    r[4] for r in data
    if (r[0] == 2 and \
        r[1] == 64 and \
        r[2] in [1000000, 5000000, 10000000, 50000000, 100000000, 500000000] and \
        r[3] == P)
]), label="Vector B+ Tree, B=64")
plt.legend(loc="lower right")
plt.xlabel("nsInterval")
plt.ylabel("Read commands (log10 scale)")
plt.title("Thoroughput for increasing nsInterval, P=256")
plt.show()
```
